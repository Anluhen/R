# Visualização de dados com ggplot2
## Motivação 
- Sintaxe do ggplot2 e scatter plot - gráficos de dispersão
- Histogramas
- Density plots - densidade de probabilidade
- Box plots - graficos de caixa
- Ajustando uma regressão dentro do ggplot2
- Plot de 2 séries e um mesmo gráfico

```r
# Carregando a biblioteca e a base de dados

library(ggplot2)

data(iris)
head(iris)
```

```r
# Scatter plot simples

# aes: são os dados dos eixos
# geom_point(): indica que vamos fazer o scatter plot
#theme_minimal(): indica o layout

ggplot(data = iris
      , aes(x = Sepal.Width, y = Sepal.Length)) + 
geom_point() +
theme_minimal()
```

```r
# Adicionando a cor baseada e classes de dados

# adicionamos o parâmetro color detro do aes

ggplot(data = iris
      , aes(x = Sepal.Width, y = Sepal.Length, color = Species)) + 
geom_point() +
theme_minimal()
```

```r
# Histograma

# Objetos ggplot podem ser armazenados em variáveis

hist <- ggplot(data = iris, aes(x = Petal.Width))
```

```r
# E ter suas componenetes adicionadas em sequência

# geom_histogram(): configura o histograma
# xlab e ylab: nomes dos eixos x e y
# ggtitle(): adiciona um título ao gráfico

hist +
geom_histogram(binwidth = 0.1, color = "black", aes(fill = Species)) +
xlab("Largura da Pétala") +
ylab("Frequência")
ggtitle("Histograma da Largura da Pétala")
```

### Density Plots - Densidade de Probabilidade
- Um gráfico de densidade é uma representação da distribuição de uma variável nmérica. É uma versão suavizada do histograma e é usada no mesmo conceito.
- São usados para estudar a distribuição de uma ou várias variáveis. Lembre-se que verificar a disribuição das sua variáveis é provavelmetne a primeira tarefa que você deve executar quando obtiver uma nova base de dados.

```r
# Calculando a média da largura por grupo de espécie
library(plyr)

media <- ddply(iris, "Species", summarise, media = mean(Petal.Width))

head(media)
```

```r
# Aplicando a média dentro do ggplot

ggplot(iris, aes(x = Petal.Width, fill = Species)) + 
geom_density(alpha = 0.3) +
geom_vline(data = media, aes(xintercept = media)
          , color = c("red", "green", "blue"), linetype = "dashed", size = 0.75) +
xlab("Largura da Pétala") +
ylab("Densidade") +
ggtitle("Density plot da largura da Pétala")
```

### Boxplot - Gráfico de Caixa
- Os gráficos de caixa incluem uma maneira simples de representar dados estatísticos, incluindo o segundo e o terceiro quartil. A linha vertical é usada para indicar o valor mediano, enquano os quartis inferior e superior são exibidos em ambos os lados do retãngulo.

```r
# Box Plot

ggplot(data = iris, aes(x = Species, y = Petal.Length, fill = Species)) +
geom_boxplot() +
theme_minimal() +
theme(legend.position = "right") +
ylab("Comprimento da Pétala") +
ggtitle("Box plot da largura da Pétala")
```

```r
# Voce pode alterar a forma do boxplot com geom_violin(), mudando também a interpretação
# é importante para entender onde temos maior concentração dos valores, algo que não
# se vê diretamente no boxplot tradicional

ggplot(data = iris, aes(x = Species, y = Petal.Width, fill = Species)) +
geom_violin() +
theme_minimal() +
theme(legend.position = "right") +
ylab("Largura da Pétala") +
ggtitle("Violin plot da largura da Pétala")
```

```r
# Repare como a "intersecção" do comprimento e largura das pétalas dificultam a classificação

ggplot(data = iris
      , aes(x = Petal.Width, y = Petal.Length, color = Species)) +
geom_point() +
theme_minimal()
```

 ### Ajustando uma regressão dentro do ggplot

```r
# Carregando as bibliotecas

library(quantmod)
library(tidyverse)
```

```r
# Capturando dados

start_date <- "2014-01-01"
end_date <- "2020-02-21"

PETR <- getSymbols("PETR4.SA", from = start_date, to = end_date, auto.assign = FALSE)
```

```r
PETR <- data.frame(PETR)

names(PETR) <- c("Open", "High", "Low", "Close", "Volume", "Ajustado")

head(PETR)
```

```r
# Calculando os retornos e o alvo

PETR <- mutate(PETR, Close_lead = lead(Close))
PETR <- mutate(PETR, Close_lag = lag(Close)) # Vai criar um NA na primeira linha

PETR <- na.omit(PETR)

PETR$Return <- (PETR$Close/PETR$Close_lag-1)*100

PETR <- mutate(PETR, Return_lag = lag(Return))

PETR <- na.omit(PETR)
```

```r
ggplot(PETR, aes(Return_lag, Return)) +
geom_point() +
geom_smooth(method = "lm") +
xlab("Lag Retorno") + ylab("Retorno") + ggtitle("Retornos PETR4")
```

```r
# Como visualizar os resíduos desta regressão?

ggplot(glm(Return ~ Return_lag, data = PETR)) +
geom_point(aes(x=.fitted, y=.resid)) + 
xlab("Ajustado") +
ylab("Resíduo") + ggtitle("Analise de Resíduos") +
theme(plot.title = element_text(hjust = 0.5))
```

```r
# E se criarmos subconjuntos?
 
PETR_sub <- subset(PETR, Return_lag > "3")
```

```r
ggplot(PETR_sub, aes(Return_lag, Return)) +
geom_point() + 
geom_smooth(method = "lm") +
xlab("Lag Retorno") + ylab("Retorno") + ggtitle("Retornos Filtrados PETR4")
```

```r
# Mas será mesmo que aquele é o melhor fit?

ggplot(PETR_sub, aes(Return_lag, Return)) +
geom_point() +
geom_smooth(method = "lm", formula = y ~ poly(x,2)) +
xlab("Lag Retorno") + ylab("Retorno") + ggtitle("Retornos Filtrados PETR4")
```

### Plot de duas sériem em um mesmo gráfico

```r
library(caTools)
```

```r
PETR$SD_Return <- c(rep(NA, 15-1), runsd(PETR$Return, 15, endrule = c("trim")))
```

```r
ggplot(PETR) + aes(Close) +
geom_smooth(aes(y = SD_Return, colour = "Volatilidade"), size = 1, alpha = 0.25) +
geom_point(aes(y = Return, colour = "Retorno")) +
geom_point(aes(y = Return_lag, colour = "Lag Retorno")) +
xlab("Preços") +
labs(color = "Elemento:") +
ylab("Retornos") + ggtitle("Retornos por preço - PETR4") +
scale_color_brewer(palette = "Set1") +
theme(plot.title = element_text(hjust = 0.5), legend.position = "bottom")
```
